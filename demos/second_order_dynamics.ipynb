# Understanding Nonlinear Dynamics

## Phase Portraits for a pendulum

```python
import numpy as np
import matplotlib.pyplot as plt
g,L = 1,1
limits = [-8, 8, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = x2_values
x2dot = -g/L*np.sin(x1_values)
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.show()
```

Now let's add damping:

```python
g,L = 1,1
limits = [-8, 8, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = x2_values
x2dot = -g/L*np.sin(x1_values) - x2_values
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.show()
```

That's about as far as we can take graphical analysis. Our next tool is linearization

## Linearization

For an equilibrium point, $p$, we can write $\dot x = f(p) + \frac{\partial f}{\partial x} |_{x=p} (x-p) + \mathrm{HOT}$

This works for vector valued $x,f$

\begin{equation}
J = \begin{bmatrix}
  \frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2}\\
  \frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2}
\end{bmatrix}_{x=p}
\end{equation}

Now $\dot y \approx Ay$, where $A=J$ and $y = x- p$. In a small neighborhood of $p$, this is a good approximation. In fact, this can tell us about the stability properties of the fixed point of the corresponding nonlinear system using linear systems theory. We must simply find the eigenvalues, $\lambda_1, \lambda_2$, for $A$. Recall that to find the eigenvalues analytically, we simply solve the system of equations

\begin{equation}
\mathrm{det}(A - \lambda I) = 0.
\end{equation}

Let's start with a situation where $\lambda_1, \lambda_2 \in \mathbb{R}$

If $\lambda_1 < 0, \lambda_2 < 0$, we have a ***stable node***.

```python
# A = [-1.0, -0.2; -0.2, -1.0]
g,L = 1,1
limits = [-3, 3, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = -1.0 * x1_values - 0.2 * x2_values
x2dot = -0.2 * x1_values - 1.0 * x2_values
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.axis('square')
plt.show()
```

If $\lambda_1 > 0, \lambda_2 > 0$, we have an ***unstable node***.

```python
# A = [1.0, -0.2; -0.2, 1.0]
g,L = 1,1
limits = [-3, 3, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = 1.0 * x1_values - 0.2 * x2_values
x2dot = -0.2 * x1_values + 1.0 * x2_values
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.axis('square')
plt.show()
```

If $\lambda_1 > 0, \lambda_2 < 0$ (or vice versa), we have a ***saddle point***.

```python
# A = [1.0, -0.2; -0.2, -1.0]
g,L = 1,1
limits = [-3, 3, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = 1.0 * x1_values - 0.2 * x2_values
x2dot = -0.2 * x1_values - 1.0 * x2_values
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.axis('square')
plt.show()
```

Now we will look at complex eigenvalues $\lambda_{1,2} = \alpha \pm j\beta$

If $\alpha < 0$, we have a ***stable focus***.

```python
# A = [0.0, 1.0; -10.0, -1.0]
g,L = 1,1
limits = [-3, 3, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = 0.0 * x1_values + 1.0 * x2_values
x2dot = -10 * x1_values - 1.0 * x2_values
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.axis('square')
plt.show()
```

Now we will look at complex eigenvalues $\lambda_{1,2} = \alpha \pm j\beta$

If $\alpha > 0$, we have an ***unstable focus***.

```python
# A = [0.0, 1.0; -10.0, 1.0]
g,L = 1,1
limits = [-3, 3, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = 0.0 * x1_values + 1.0 * x2_values
x2dot = -10 * x1_values + 1.0 * x2_values
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.axis('square')
plt.show()
```

Now we will look at complex eigenvalues $\lambda_{1,2} = \alpha \pm j\beta$

If $\alpha = 0$, we have a ***center***.

```python
# A = [0.0, 10.0; -10.0, -1.0]
g,L = 1,1
limits = [-3, 3, -3, 3]
x1_values, x2_values = np.meshgrid(np.arange(limits[0], limits[1], 0.1), np.arange(limits[2], limits[3], 0.1))
x1dot = 0.0 * x1_values + 10.0 * x2_values
x2dot = -10 * x1_values - 1.0 * x2_values
plt.streamplot(x1_values, x2_values, x1dot, x2dot, color='r', linewidth=0.5, density=1.6)
plt.axis(limits)
plt.axis('square')
plt.show()
```
